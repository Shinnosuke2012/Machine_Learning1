## 数値と四則演算

「'」や「"」で括らないと，数値や変数などとして扱われます．

```
print(8)
```

print関数の中に，直接，計算式を書くこともできます．

```
print(5+3)
```

```
print(333 + 5 * 111)
```

上の例で，「*」 は掛け算（乗算）の演算子を表すので，通常の数式で表すと $333 + 5 \times 111$ を計算していることになります．四則演算の規則通りに，足し算よりも掛け算を先に実行していることに注意しましょう．

割り算（除算）に関しては少し注意が必要です．CやJava等を始めとする他の多くのプログラム言語と同様に，割り算を「/」で表しますが，

```
print(32 / 5)
```

整数と整数の除算の結果の商（これも整数）を求めたいときは，「//」を使います．

```
print(32 // 5)
```

また，整数の除算による余りを求めるには，「%」を使います．

```
print(32 % 5)
```

計算式の中に括弧 ( ) を使うことができます．

```
print((1+2)*(2+3)-7)
```

実は，Colaboratory のセルでは最後に評価した式を表示する仕組みがあるので，print関数を明示的に書かなくても計算結果を知ることはできます．

## 変数

「変数」に数値や文字列などを「代入」して蓄えることができます．

```
x = 12
y = 76
z = x + y 
print("x = ", x)
print("y = ", y)
print("z = ", z)
```

表示を見やすくするために，色々な書き方があります．

* カンマで区切って並べていく方法．



```
print("xの値は", x, "で，yの値は", y, "です．そしてzの値は", z, "です．")
```

* 文字列の連結を使う方法．

```
print("xの値は" + str(x)+ "で，yの値は" + str(y) + "です．そしてzの値は" +str(z)+ "です．")
```

* **%記法**．C言語の printf関数の書き方に近い．

```
print("xの値は%dで，yの値は%dです．そしてzの値は%dです．" % (x, y, z))
```

* **format記法**．

```
print("xの値は{0}で，yの値は{1}です．そしてzの値は{2}です．".format(x, y, z))
```

* **f文字列** *（フォーマット済み文字列リテラル)*．先頭の「f」に注目．



```
print(f"xの値は{x}で，yの値は{y}です．そしてzの値は{z}です．")
```

このように，様々な書き方があるので，好みに応じて使い分ければよいですが，個人的には最後の**f文字列**がお勧めです．直感的にわかりやすいし，プログラムを修正する際にミスが発生しにくいと感じます．

【練習課題】ここで示した機能を使って，何かを表示してみましょう．

去年線形代数Bでグラムシュミットの直交基底について学習した。正規直交基底を求めることで、直交行列の生成が可能となり、対角化をより容易にする。誰かにシェアしたかったので、今ここで使いたいと思う。

```
#encoding:utf-8
import numpy as np

def gram_schmit(a_1,a_2,a_3):
  #直交化したいベクトルを引数にとる
  u_1 = a_1 / np.linalg.norm(a_1) #u_1の計算
  c_1 = np.dot(a_2,u_1)
  u_2 = a_2 - c_1 * u_1 / np.linalg.norm(a_2 - c_1 * u_1)
  c_2 = np.dot(a_3,u_2)
  u_3 = a_3 - c_1 * u_1 - c_2 * u_2 / np.linalg.norm(a_3 - c_1 * u_1 - c_2 * u_2)

  print("Value of constants")
  print(f'c_1: {c_1}\nc_2: {c_2}')
  print("\n")
  print("Value of vectors")
  print(f'u_1:{u_1}\nu_2:{u_2}\nu_3:{u_3}')

#(1,0,1) (1,1,0) (0,1,1)をグラムシュミットの直交化を用いて求めてみる
a_1 = np.array([1,0,1], dtype='float')
a_2 = np.array([1,1,0], dtype='float')
a_3 = np.array([0,1,1], dtype='float')

gram_schmit(a_1,a_2,a_3)
```
